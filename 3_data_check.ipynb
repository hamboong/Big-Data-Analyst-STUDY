## **데이터 조회**
- Pandas 함수를 통해 원하는 조건의 데이터 조회를 할 수 있다.
- 정렬, 조건 함수, 인덱스 등을 활용한다.
- loc, iloc 함수를 통해 인덱스 기반 접근이 가능하다.

## **Data : Titanic 생존자 데이터 분석**

- 1912년 발생한 타이타닉호 침몰 사건 데이터를 활용한다.
- Kaggle 에서 데이터를 다운받을 수 있다.(https://www.kaggle.com/c/titanic/data?select=train.csv)

## **데이터 설명**
- survival: Survival (0 = No; 1 = Yes)
- pclass: Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
- name: Name
- sex: Sex
- age: Age
- sibsp: Number of Siblings/Spouses Aboard
- parch: Number of Parents/Children Aboard
- ticket: Ticket Number
- fare: Passenger Fare
- cabin: Cabin
- embarked: Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

## **loc인덱싱**
